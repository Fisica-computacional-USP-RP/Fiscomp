## Grafo completo

```python
import numpy as np

def grafo_completo(N):
    '''gera a matriz de adjacência de um grafo grafo completo '''

    adj = np.ones((N,N),dtype=int)                # matriz de 1's
    for i in range(N): adj[i][i] = 0    # tentar escrever de forma concisa
    return adj


N = 100
A = grafo_completo(N)
print("A = ",A)

```

## Cadeia

```python
import numpy as np

def grafo_cadeia(N,cp):
    ''' gera a matriz de adjacência de uma cadeia '''
    ''' com (sem) cp = 1 (0) condição periódica de contorno'''

    adj = np.zeros((N,N),dtype=int) # matriz NxN de zeros
    for i in range(N):              # laço sobre os nodos
        i1 = i + 1       # nó posterior ao nó considerado
        if i1 < N:
            adj[i ][i1] = 1
            adj[i1][i ] = 1
    if cp ==  1:
        N1 = N-1
        adj[0][N1] = 1
        adj[N1][0] = 1
    return adj

N  = 10
cp = 1
A = grafo_cadeia(N,cp)
print("A = ",A)

G = nx.from_numpy_array(np.matrix(A))
nx.draw_circular(G)
```

## Cadeia $k$ vizinhos



```python
import numpy as np
import networkx as nx # (para visualização do grafo)


def grafo_cadeia_k_vizinhos(N,K,CP):
  '''gera a matriz de adjacência de uma cadeia com N nodos com K < N vizinhos'''
  ''' sem (com) CP = 0 (1) condição periódica de contorno'''
  if (K >= N): print("erro rotina grafo cadeia_k_vizinhos")
  adj = np.zeros((N,N),dtype=int)
  for i in range(N):
    i0 = i + 1
    for j in range(K):
      i1 = i0 + j
      if i1 < N:
        adj[i ][i1] = 1
        adj[i1][i ] = 1
  if cp ==  1:
    N1 = N - 1
    for i in range(K):
      for j in range(K-i):
        i1 = N1 - j
        if i1 < N:
          adj[i ][i1] = 1
          adj[i1][i ] = 1
  return adj

N  = 10
K  = 2
cp = 1
A  = grafo_cadeia_k_vizinhos(N,K,cp)
print("A = ",A)

G = nx.from_numpy_array(np.matrix(A))
nx.draw_circular(G)
```

# Grafo de Erdos-Renyi

```python
import numpy as np
import networkx as nx

def grafo_aleatorio_ErdosRenyi(N,L):
    '''gera a matriz de adjacência de um grafo grafo aleatório '''
    '''onde L é número de ligações '''
    adj = np.zeros((N,N),dtype=int)                        # matriz NxN de zeros
    for k in range(L):                     # laço para varredura de todos os nós
      i = np.random.randint(0,N)
#      print("i = ",i)
      j = np.random.randint(0,N)           # existe a possibilidade de autoligação
#      print("j = ",j)
#      print("i,j = ",i,j)
      adj[i][j] = 1
      adj[j][i] = 1
    return adj

N  =  6
L  = 8                                 # prob. de dois nós estarem ligados é: p = L/[N(N-1)/2]
A  = grafo_aleatorio_ErdosRenyi(N,L)
# print("A = ",A)
G = nx.from_numpy_array(np.matrix(A))
# nx.draw(G)


#g1 = nx.gnp_random_graph(N, p, seed=None, directed=False) # transforma matriz de adjacencia em um grafo
# nx.draw(G)
nx.draw_circular(G)
```

# Comparação entre grafos de Gilbert e Erdos-Renyi

# Comparação entre número de ligações calculadas de grafos aleatórios e distribuição teórica

Considere $n$ nodos e a probabilidade de ligação $p$ entre dois nodos diferentes (o número total de ligações é $n(n-1)/2$).
A distribuição do número de ligações $L$ é binomial:
$$
B_{n(n-1)/2,p}(L) = \binom{\frac{n(n-1)}{2}}{L} p^L (1-p)^{\frac{n(n-1)}{2} - L} \; .
$$

```python
# código para gera m grafos aleatorios, mostrar histograma do numero de ligacoes
# e comparar com a distribuição binomial (distribuição teórica)

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

np.random.seed(seed=None)
n = 20   # número de nodos
p = .4   # probabilidade de sucesso
m = 1000  # numero de experimentos binomiais

def grafo_aleatorio(N,p):
    '''gera a matriz de adjacência de um grafo grafo aleatório '''
    '''onde dois nodos se conectam com prob. p '''
    adj = np.zeros((N,N),dtype=int)                        # matriz NxN de zeros
    for i in range(N):                     # laço para varredura de todos os nós
        i1 = i + 1
        for j in range(i1,N):  # laço para varredura dos nós seguintes do nó considerado
          if np.random.rand() < p:
            adj[i][j] = 1
            adj[j][i] = 1
    return adj

N  = int(n*(n-1)/2)  # numero de ligações, número de experimento de Bernoulli
# print ("N = ",N)

x = np.arange(N+1)
# print("x = ",x)

res = np.zeros(m,dtype=int)              # lista de resultados (todos fracassos)
for j in range(m):
  A  = grafo_aleatorio(n,p)
  L  = np.sum(A)/2
#  print("A = ",A)
#  print("L = ",L)
  res[j] = L
#print("res = ", res)
fig3 = plt.hist(res,bins=x,density=True)   # histograma de L dos grafos gerados


y = st.binom.pmf(x, N, p)                                        # valor teorico
# print("y = ",y)
fig4 = plt.plot(x,y, 'bo', ms=6)                          # distribuicao teorica

# G = nx.from_numpy_array(np.matrix(A), create_using=nx.DiGraph)
# nx.draw(G)
```

# Comparação entre graus de um nodo de grafos aleatórios e distribuição teórica

Considere $n$ nodos e a probabilidade de ligação $p$ entre dois nodos diferentes (o número total de ligações é $n(n-1)/2$).
A distribuição do número de ligações $L$ é binomial:
$$
B_{n-1,p}(k) = \binom{n-1}{k} p^k (1-p)^{(n-1) - k} \; .
$$

```python
# código para gera m grafos aleatorios, mostrar histograma do numero de ligacoes
# e comparar com a distribuição binomial (distribuição teórica)

# LICAO DE CASA: MODIFICAR PARA GRAUS

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

np.random.seed(seed=None)
n = 20   # número de nodos
p = .4   # probabilidade de sucesso
m = 10000  # numero de experimentos binomiais

def grafo_aleatorio(N,p):
    '''gera a matriz de adjacência de um grafo grafo aleatório '''
    '''onde dois nodos se conectam com prob. p '''
    adj = np.zeros((N,N),dtype=int)                        # matriz NxN de zeros
    for i in range(N):                     # laço para varredura de todos os nós
        i1 = i + 1
        for j in range(i1,N):  # laço para varredura dos nós seguintes do nó considerado
          if np.random.rand() < p:
            adj[i][j] = 1
            adj[j][i] = 1
    return adj

N  = int(n*(n-1)/2)  # numero de ligações, número de experimento de Bernoulli
# print ("N = ",N)

x = np.arange(N+1)
# print("x = ",x)

res = np.zeros(m,dtype=int)              # lista de resultados (todos fracassos)
for j in range(m):
  A  = grafo_aleatorio(n,p)
  L  = np.sum(A)/2
#  print("A = ",A)
#  print("L = ",L)
  res[j] = L
#print("res = ", res)
fig3 = plt.hist(res,bins=x,density=True)   # histograma de L dos grafos gerados


y = st.binom.pmf(x, N, p)                                        # valor teorico
# print("y = ",y)
fig4 = plt.plot(x,y, 'bo', ms=6)                          # distribuicao teorica

# G = nx.from_numpy_array(np.matrix(A), create_using=nx.DiGraph)
# nx.draw(G)
```

```python
def grafo_rede_retangular(Lx,Ly,cp):
    '''gera uma rede retangular de arestas de Lx e Ly sítios'''
    '''com (sem) cp = 1 (0) cond. per. de contorno'''
    N    = Lx*Ly                                     # numero de sitios do grafo
    adj  = np.zeros((N,N),dtype=int)                 # cria matriz de adjacencia
    for il in range(Ly):                                       # índice de linha
      il_m = il - 1                                         # vizinho à esquerda
      il_p = il + 1                                          # vizinho à direita
      if cp == 1:                            # se condição periódica de contorno
        if il_m == -1: il_m = Ly - 1        # liga o primeiro sítio com o último
        if il_p == Ly: il_p = 0             # liga o último sítio com o primeiro
      il_Lm = il_m*Lx                        # número do sítio anterior no grafo
      il_L  = il*Lx                        # número do sítio referência no grafo
      il_Lp = il_p*Lx                       # número do sítio posterior no grafo
      for ic in range(Lx):                                    # índice de coluna
        ic_m = ic - 1                                            # vizinho acima
        ic_p = ic + 1                                           # vizinho abaixo
        if cp == 1:                          # se condição periódica de contorno
          if ic_m == -1: ic_m = Lx - 1      # liga o primeiro sítio com o último
          if ic_p == Lx: ic_p = 0           # liga o último sítio com o primeiro
        ku   = il_Lm + ic                                          # sítio acima
        km   = il_L  + ic_m                                   # sítio à esquerda
        k    = il_L  + ic                                  # sítio de referência
        kp   = il_L  + ic_p                                    # sítio à direita
        kd   = il_Lp + ic                                         # sítio abaixo
#        print (ku,km,k,kp,kd),
        if ku >= 0: adj[k][ku] = 1          # liga sítio de ref. com sítio acima
        if km >= 0: adj[k][km] = 1     # liga sítio de ref. com sítio à esquerda
        if kp <  N: adj[k][kp] = 1      # liga sítio de ref. com sítio à direita
        if kd <  N: adj[k][kd] = 1         # liga sítio de ref. com sítio abaixo
#        print(il,ic,ku,km,k,kp,kd)
    return adj
```

# Evolução de redes aleatórias

Reproduzimos aqui o gráfico da Fig. 3.7 do livro Network Science do Barabási. Além de reproduzir o resultado colocamos as barras de erros.

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

g     = None # cria um grafo
N     = 10000  # número de nodos do grafo
iters = 10  # número de realizações para cada valor de probabilidade p
X = []       # inicializa lista
Y = []       # inicializa lista
Y_err = []   # inicializa lista

for p in np.linspace(0.0001, .007 , 100):              # laço na probabilidade p
  # valores maiores de p não são interessantes pois a rede gerada já possui uma componente

  # print('probabilidade de conexão entre vértices %.2f' % p)

  k = p*(N-1)         # o grau médio
  # print('grau médio estimado %.2f' % k)

  N_g_configs = []  # inicializa lista
  k_configs   = []  # inicializa lista
  for _ in range(iters):                                    # laço das iterações
    g = nx.erdos_renyi_graph(N, p)                            # grafo de Gilbert

    k = np.mean(list(dict(g.degree()).values()))     # grau médio da rede gerada
    k_configs.append(k)

    giant = sorted(nx.connected_components(g), key=len, reverse=True)[0]
    N_g = len(giant)                               # tamanho do maior componente
    N_g_configs.append(N_g)

# saido do laço são duas listas: <k> e N_g

  N_g_configs_normed = np.asarray(N_g_configs)/N

  X.append(np.mean(k_configs))                    # absicissa é <k>
  Y.append(np.mean(N_g_configs_normed))           # ordenada é N_G/N
  Y_err.append(np.std(N_g_configs_normed)/np.sqrt(iters))        # desvio-padrão
                                                  # colocar o erro-padrão
plt.figure(figsize=(16,9))
plt.errorbar(X,Y,yerr=Y_err)
plt.axvline(1, c='red')
plt.axvline(np.log(N), c='red')
plt.xlabel('<k>', size=14)
plt.ylabel(r'$N_g/N$', size=14)
plt.show()
```

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

# ── Semente ───────────────────────────────────────────────────────────────────
np.random.seed(42)

# ── Parâmetros globais ────────────────────────────────────────────────────────
N:     int   = 10_000
iters: int   = 10
p_min: float = 0.0001
p_max: float = 0.007
n_pts: int   = 100

# ── Eixo de probabilidades ────────────────────────────────────────────────────
p_vals: np.ndarray = np.linspace(p_min, p_max, n_pts)   # shape (n_pts,)

# ── Arrays de saída — pré-alocados ───────────────────────────────────────────
X:     np.ndarray = np.zeros(n_pts)   # ⟨k⟩ por valor de p
Y:     np.ndarray = np.zeros(n_pts)   # ⟨N_g/N⟩ por valor de p
Y_err: np.ndarray = np.zeros(n_pts)   # erro padrão de N_g/N

# ── Arrays de realização — pré-alocados ──────────────────────────────────────
k_configs:  np.ndarray = np.zeros(iters)   # grau médio por realização
Ng_configs: np.ndarray = np.zeros(iters)   # N_g/N por realização

# ── Variáveis escalares do laço ───────────────────────────────────────────────
i:     int        # índice do laço externo
p:     float      # probabilidade de conexão atual
g:     nx.Graph   # grafo de Gilbert G(N, p)
k_obs: float      # grau médio observado
N_g:   int        # tamanho da componente gigante

# ─────────────────────────────────────────────────────────────────────────────

def gerar_grafo(N: int, p: float) -> nx.Graph:
    return nx.fast_gnp_random_graph(N, p)

def grau_medio(g: nx.Graph) -> float:
    return 2 * g.number_of_edges() / g.number_of_nodes()

def componente_gigante(g: nx.Graph) -> int:
    return len(max(nx.connected_components(g), key=len))

# ─────────────────────────────────────────────────────────────────────────────

for i, p in enumerate(p_vals):

    for j in range(iters):
        g      = gerar_grafo(N, p)
        k_obs  = grau_medio(g)
        N_g    = componente_gigante(g)

        k_configs[j]  = k_obs
        Ng_configs[j] = N_g / N

    X[i]     = np.mean(k_configs)
    Y[i]     = np.mean(Ng_configs)
    Y_err[i] = np.std(Ng_configs) / np.sqrt(iters)

# ── Visualização ──────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(16, 9))

ax.errorbar(X, Y, yerr=Y_err, fmt='-o', markersize=3,
            capsize=3, label=r'$\langle N_g/N \rangle$')
ax.axvline(1,         color='red', linestyle='--', label=r'$\langle k \rangle = 1$')
ax.axvline(np.log(N), color='red', linestyle=':',  label=r'$\langle k \rangle = \ln N$')

ax.set_xlabel(r'$\langle k \rangle$', fontsize=14)
ax.set_ylabel(r'$N_g / N$',           fontsize=14)
ax.legend(fontsize=12)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import os # Necessário para lidar com nomes de arquivos

def grafo_topologico(Lx, Ly, BC_x, BC_y, MN=0):
    '''Gera matriz de adjacência para diversas topologias 2D.'''
    N = Lx * Ly
    adj = np.zeros((N, N), dtype=int)

    def resolve_vizinho(ic, il, dic, dil):
        nc = ic + dic
        nl = il + dil

        if nc < 0 or nc >= Lx:
            if BC_x == 0: return None
            nc = nc % Lx
            if BC_x == 2: nl = Ly - 1 - nl

        if nl < 0 or nl >= Ly:
            if BC_y == 0: return None
            nl = nl % Ly
            if BC_y == 2: nc = Lx - 1 - nc

        return int(nl * Lx + nc)

    direcoes = [(0, -1), (0, 1), (-1, 0), (1, 0)]
    if MN == 1:
        direcoes += [(-1, -1), (1, -1), (-1, 1), (1, 1)]

    for il in range(Ly):
        for ic in range(Lx):
            k = il * Lx + ic
            for dic, dil in direcoes:
                vizinho_k = resolve_vizinho(ic, il, dic, dil)
                if vizinho_k is not None:
                    adj[k][vizinho_k] = 1

    return adj

def visualizar_topologia(Lx, Ly, adj, titulo):
    '''Desenha o grafo na tela destacando as conexões topológicas.'''
    G = nx.from_numpy_array(adj)

    # Posiciona os nós como uma grade retangular clássica (x, y)
    pos = {i: (i % Lx, Ly - 1 - (i // Lx)) for i in range(Lx * Ly)}

    plt.figure(figsize=(7, 6))
    plt.title(titulo, fontsize=14, fontweight='bold')

    # Desenha os nós
    nx.draw_networkx_nodes(G, pos, node_color='lightblue', edgecolors='black', node_size=600)
    nx.draw_networkx_labels(G, pos, font_weight='bold')

    # Separa as arestas locais das arestas "topológicas" (que dão a volta/torcem)
    arestas_locais = []
    arestas_topologicas = []

    for u, v in G.edges():
        x1, y1 = pos[u]
        x2, y2 = pos[v]
        # Se a distância na tela for maior que 1 (em x ou y) e maior que a diagonal (sqrt(2) ~ 1.41)
        # Significa que é uma conexão que atravessa as bordas
        dist = np.sqrt((x1 - x2)**2 + (y1 - y2)**2)
        if dist > 1.5:
            arestas_topologicas.append((u, v))
        else:
            arestas_locais.append((u, v))

    # Arestas normais da malha (retas e pretas)
    nx.draw_networkx_edges(G, pos, edgelist=arestas_locais, width=2.0, alpha=0.7)

    # Arestas topológicas (curvas e vermelhas) para mostrar o "salto"
    nx.draw_networkx_edges(G, pos, edgelist=arestas_topologicas, width=2.0,
                           alpha=0.6, edge_color='red', connectionstyle='arc3,rad=0.3')

    plt.axis('off')
    plt.tight_layout()
    plt.show()

# ==============================================================================
# PAINEL DE EXECUÇÃO E EXPORTAÇÃO
# ==============================================================================

if __name__ == "__main__":
    # Dimensões da malha
    Lx, Ly = 4, 4
    MN = 1  # 0 = von Neumann, 1 = Moore

    # Pasta onde os arquivos serão salvos (cria se não existir)
    pasta_saida = "matrizes_adjacencia"
    if not os.path.exists(pasta_saida):
        os.makedirs(pasta_saida)

    # Dicionário com todas as configurações
    topologias = {
        "Plano_Finito": (0, 0),
        "Cilindro": (1, 0),
        "Toro": (1, 1),
        "Mobius": (2, 0),
        "Garrafa_Klein": (2, 1),
        "Plano_Projetivo": (2, 2)
    }

    print(f"Gerando malhas {Lx}x{Ly} com Vizinhança = {'Moore (8)' if MN == 1 else 'von Neumann (4)'}\n")

    for nome, (bc_x, bc_y) in topologias.items():
        print(f"Processando: {nome}...")

        # 1. Gera a matriz
        matriz_adj = grafo_topologico(Lx, Ly, bc_x, bc_y, MN)

        # 2. Define o nome do arquivo (ex: Toro_MN1_4x4.csv)
        nome_arquivo = f"{nome}_MN{MN}_{Lx}x{Ly}.csv"
        caminho_completo = os.path.join(pasta_saida, nome_arquivo)

        # 3. Salva no arquivo (formato inteiro '%d', separado por vírgula)
        np.savetxt(caminho_completo, matriz_adj, fmt='%d', delimiter=',')

        print(f" -> Salvo com sucesso em: {caminho_completo}")

        # 4. (Opcional) Visualiza - comente a linha abaixo se quiser apenas salvar
        # visualizar_topologia(Lx, Ly, matriz_adj, nome.replace('_', ' '))

    print("\nTodos os arquivos foram gerados!")
```

Gera redes bidimensionais quadradas e salva arquivo em forma de lista de ligações

```python
import numpy as np
import os

def lista_arestas_topologica(Lx, Ly, BC_x, BC_y, MN=0):
    '''
    Gera uma Lista de Arestas (Edge List) para diversas topologias 2D.
    Ideal para malhas grandes (L > 50).
    '''
    arestas = []

    def resolve_vizinho(ic, il, dic, dil):
        nc = ic + dic
        nl = il + dil

        if nc < 0 or nc >= Lx:
            if BC_x == 0: return None
            nc = nc % Lx
            if BC_x == 2: nl = Ly - 1 - nl

        if nl < 0 or nl >= Ly:
            if BC_y == 0: return None
            nl = nl % Ly
            if BC_y == 2: nc = Lx - 1 - nc

        return int(nl * Lx + nc)

    direcoes = [(0, -1), (0, 1), (-1, 0), (1, 0)]
    if MN == 1:
        direcoes += [(-1, -1), (1, -1), (-1, 1), (1, 1)]

    for il in range(Ly):
        for ic in range(Lx):
            k = il * Lx + ic

            for dic, dil in direcoes:
                vizinho_k = resolve_vizinho(ic, il, dic, dil)

                # Registra a aresta apenas se o vizinho existe
                if vizinho_k is not None:
                    # Para evitar arestas duplicadas (ex: 0 ligando ao 1 e 1 ligando ao 0)
                    # em grafos não-direcionados, salvamos apenas quando k < vizinho_k
                    if k < vizinho_k:
                        arestas.append((k, vizinho_k))

    return np.array(arestas)

# ==============================================================================
# PAINEL DE EXECUÇÃO E EXPORTAÇÃO
# ==============================================================================

if __name__ == "__main__":
    # Agora você pode colocar valores grandes sem travar o computador!
    Lx, Ly = 100, 100  # Malha de 10.000 nós
    MN = 1             # 1 = Moore (8 vizinhos)

    pasta_saida = "listas_arestas"
    if not os.path.exists(pasta_saida):
        os.makedirs(pasta_saida)

    topologias = {
        "Plano_Finito": (0, 0),
        "Cilindro": (1, 0),
        "Toro": (1, 1),
        "Mobius": (2, 0),
        "Garrafa_Klein": (2, 1),
        "Plano_Projetivo": (2, 2)
    }

    print(f"Gerando malhas {Lx}x{Ly} com Vizinhança = {'Moore' if MN == 1 else 'von Neumann'}\n")

    for nome, (bc_x, bc_y) in topologias.items():
        print(f"Processando: {nome}...")

        # Gera a lista de arestas diretamente
        arestas = lista_arestas_topologica(Lx, Ly, bc_x, bc_y, MN)

        nome_arquivo = f"{nome}_arestas_MN{MN}_{Lx}x{Ly}.txt"
        caminho_completo = os.path.join(pasta_saida, nome_arquivo)

        # Salva o arquivo em duas colunas separadas por espaço (padrão em softwares de redes)
        np.savetxt(caminho_completo, arestas, fmt='%d', delimiter=' ',
                   header='Source Target', comments='')

        print(f" -> Salvo com sucesso! ({len(arestas)} arestas geradas)")

    print("\nTodos os arquivos foram gerados!")
```

Redes bidimensionais hexagonais:
A rede hexagonal é um grafo bipartido (ela pode ser dividida em duas sub-redes, A e B, onde os nós de A só se conectam com os nós de B). No nosso código, isso é garantido pela regra da paridade (il + ic) % 2.

Por causa dessa natureza, ao usar Condições Periódicas (1) ou Torcidas (2), as dimensões Lx e Ly devem obrigatoriamente ser números pares.

Se você usar um número ímpar (ex: uma malha 51×51) e tentar fechar a malha em um Toro ou Cilindro, a "costura" da borda conectará dois nós que apontam para a mesma direção vertical. Isso criaria defeitos topológicos na malha (o equivalente a adicionar "pentágonos" ou "heptágonos" na estrutura perfeita de grafeno). Mantendo dimensões pares, a malha se fecha perfeitamente sem criar frustrações na vizinhança.

```python
import numpy as np
import os

def lista_arestas_hexagonal(Lx, Ly, BC_x, BC_y):
    '''
    Gera uma Lista de Arestas para uma rede hexagonal (favo de mel) 2D.
    Utiliza o mapeamento topológico de "Parede de Tijolos" (Brick Wall).
    Cada nó possui exatamente 3 primeiros vizinhos no bulk.
    '''
    arestas = []

    def resolve_vizinho(ic, il, dic, dil):
        nc = ic + dic
        nl = il + dil

        # Avalia cruzamento no eixo X (Colunas)
        if nc < 0 or nc >= Lx:
            if BC_x == 0: return None
            nc = nc % Lx
            if BC_x == 2: nl = Ly - 1 - nl

        # Avalia cruzamento no eixo Y (Linhas)
        if nl < 0 or nl >= Ly:
            if BC_y == 0: return None
            nl = nl % Ly
            if BC_y == 2: nc = Lx - 1 - nc

        return int(nl * Lx + nc)

    for il in range(Ly):
        for ic in range(Lx):
            k = il * Lx + ic

            # 1. Todo nó sempre se conecta na horizontal (Leste e Oeste)
            # (dic, dil) -> variação na (coluna, linha)
            direcoes = [(-1, 0), (1, 0)]

            # 2. A conexão vertical alterna dependendo da paridade (x + y)
            if (il + ic) % 2 == 0:
                direcoes.append((0, 1))  # Conecta para Baixo (Sul)
            else:
                direcoes.append((0, -1)) # Conecta para Cima (Norte)

            for dic, dil in direcoes:
                vizinho_k = resolve_vizinho(ic, il, dic, dil)

                # Se o vizinho existe e não é uma borda aberta
                if vizinho_k is not None:
                    # Salva apenas uma vez cada par (grafo não-direcionado)
                    if k < vizinho_k:
                        arestas.append((k, vizinho_k))

    return np.array(arestas)

# ==============================================================================
# PAINEL DE EXECUÇÃO E EXPORTAÇÃO
# ==============================================================================

if __name__ == "__main__":
    # IMPORTANTE: Para redes hexagonais com condições periódicas/torcidas,
    # Lx e Ly DEVEM ser números PARES para evitar "frustração" topológica.
    Lx, Ly = 100, 100

    pasta_saida = "listas_arestas_hex"
    if not os.path.exists(pasta_saida):
        os.makedirs(pasta_saida)

    topologias = {
        "Plano_Hexagonal": (0, 0),
        "Cilindro_Hexagonal": (1, 0),
        "Toro_Hexagonal": (1, 1),
        "Mobius_Hexagonal": (2, 0),
        "Garrafa_Klein_Hexagonal": (2, 1)
    }

    print(f"Gerando malhas hexagonais (Brick Wall) {Lx}x{Ly}\n")

    for nome, (bc_x, bc_y) in topologias.items():
        print(f"Processando: {nome}...")

        arestas = lista_arestas_hexagonal(Lx, Ly, bc_x, bc_y)

        nome_arquivo = f"{nome}_{Lx}x{Ly}.txt"
        caminho_completo = os.path.join(pasta_saida, nome_arquivo)

        np.savetxt(caminho_completo, arestas, fmt='%d', delimiter=' ',
                   header='Source Target', comments='')

        print(f" -> Salvo com sucesso! ({len(arestas)} arestas geradas)")

    print("\nTodos os arquivos foram gerados!")
```

```python
Redes cúbicas:

O que cada "Contorno" representa fisicamente?

    Toro 3D (1, 1, 1): Esta é a configuração mais usada em física da matéria condensada. Se uma partícula (ou interação) sai pela face direita do cubo, ela reentra instantaneamente pela face esquerda. Isso simula um material infinito, eliminando os "efeitos de superfície" onde os átomos das bordas têm menos vizinhos.

    Placa Infinita (1, 1, 0): Muito útil para simular filmes finos magnéticos ou grafite. Os eixos X e Y se repetem ao infinito, mas o eixo Z representa a espessura finita real do material (tem um topo e um fundo físicos).
```

```python
import numpy as np
import os

def lista_arestas_cubica(Lx, Ly, Lz, BC_x, BC_y, BC_z):
    '''
    Gera Lista de Arestas para rede cúbica simples 3D (6 primeiros vizinhos).
    BC = 0 (Aberta), 1 (Periódica), 2 (Torcida/Invertida).
    '''
    arestas = []

    # Constante de área para facilitar o cálculo do ID do nó (k)
    Area = Lx * Ly

    def resolve_vizinho(x, y, z, dx, dy, dz):
        nx = x + dx
        ny = y + dy
        nz = z + dz

        # Avalia cruzamento no eixo X (Esquerda/Direita)
        if nx < 0 or nx >= Lx:
            if BC_x == 0: return None
            nx = nx % Lx
            if BC_x == 2:
                ny = Ly - 1 - ny
                nz = Lz - 1 - nz

        # Avalia cruzamento no eixo Y (Cima/Baixo)
        if ny < 0 or ny >= Ly:
            if BC_y == 0: return None
            ny = ny % Ly
            if BC_y == 2:
                nx = Lx - 1 - nx
                nz = Lz - 1 - nz

        # Avalia cruzamento no eixo Z (Frente/Trás)
        if nz < 0 or nz >= Lz:
            if BC_z == 0: return None
            nz = nz % Lz
            if BC_z == 2:
                nx = Lx - 1 - nx
                ny = Ly - 1 - ny

        return int(nz * Area + ny * Lx + nx)

    # 6 direções de primeiros vizinhos em 3D
    direcoes = [
        (1, 0, 0), (-1, 0, 0),  # +/- X
        (0, 1, 0), (0, -1, 0),  # +/- Y
        (0, 0, 1), (0, 0, -1)   # +/- Z
    ]

    for z in range(Lz):
        for y in range(Ly):
            for x in range(Lx):
                k = z * Area + y * Lx + x

                for dx, dy, dz in direcoes:
                    vizinho_k = resolve_vizinho(x, y, z, dx, dy, dz)

                    if vizinho_k is not None:
                        # Salva apenas quando k < vizinho_k (evita arestas duplicadas)
                        if k < vizinho_k:
                            arestas.append((k, vizinho_k))

    return np.array(arestas)

# ==============================================================================
# PAINEL DE EXECUÇÃO E EXPORTAÇÃO
# ==============================================================================

if __name__ == "__main__":
    Lx, Ly, Lz = 20, 20, 20  # Malha de 8.000 nós

    pasta_saida = "listas_arestas_3D"
    if not os.path.exists(pasta_saida):
        os.makedirs(pasta_saida)

    # Definindo as topologias em 3 eixos: (BC_x, BC_y, BC_z)
    topologias = {
        "Cubo_Finito": (0, 0, 0),                # Todas as bordas fechadas/rígidas
        "Fio_Infinito": (0, 0, 1),               # Apenas Z é periódico (cabo longo)
        "Placa_Infinita": (1, 1, 0),             # X e Y periódicos, Z finito (filme fino)
        "Toro_3D": (1, 1, 1),                    # "Universo" periódico em todas as direções
        "Cubo_Mobius_X": (2, 0, 0),              # Sólido torcido no eixo X
        "Sólido_Projetivo_3D": (2, 2, 2)         # Todas as dimensões sofrem torção ao cruzar
    }

    print(f"Gerando malhas cúbicas 3D: {Lx}x{Ly}x{Lz} nós\n")

    for nome, (bc_x, bc_y, bc_z) in topologias.items():
        print(f"Processando: {nome}...")

        arestas = lista_arestas_cubica(Lx, Ly, Lz, bc_x, bc_y, bc_z)

        nome_arquivo = f"{nome}_{Lx}x{Ly}x{Lz}.txt"
        caminho_completo = os.path.join(pasta_saida, nome_arquivo)

        np.savetxt(caminho_completo, arestas, fmt='%d', delimiter=' ',
                   header='Source Target', comments='')

        print(f" -> Salvo! ({len(arestas)} arestas geradas)")

    print("\nConcluído!")
```

O modelo de Potts (uma generalização do modelo de Ising para q estados) é o laboratório perfeito para testar como a geometria e a topologia afetam as transições de fase. Aplicar esse modelo nas estruturas que acabamos de gerar abre portas para análises muito ricas.

Aqui estão os principais motivos pelos quais essa abordagem é tão poderosa:
1. Eliminação de Efeitos de Superfície (Scaling de Tamanho Finito)

Em uma malha com bordas abertas, os spins da superfície têm menos vizinhos (coordenação menor) do que os spins do núcleo. Isso cria um comportamento termodinâmico artificial, pois os spins da borda "congelam" ou "derretem" em temperaturas diferentes do resto do material. Ao usar o Toro (Condições Periódicas em tudo), todos os nós passam a ter exatamente o mesmo número de vizinhos. Isso permite simular um material "infinito" usando apenas alguns milhares de nós, tornando o cálculo das propriedades críticas (como a temperatura crítica Tc​) muito mais preciso.
2. Estudo de Frustração Topológica

Quando você aplica o modelo de Potts em uma Fita de Möbius ou Garrafa de Klein, a torção da fronteira pode forçar a criação de um "defeito" macroscópico no sistema. Se você estiver simulando um acoplamento antiferromagnético, por exemplo, a torção torna impossível que o sistema alcance um estado fundamental perfeito, forçando a existência permanente de uma parede de domínio (domain wall). Estudar a energia e a dinâmica dessas paredes é um tema super quente.
3. Teste de Classes de Universalidade

A topologia local (como mudar de uma rede Quadrada de 4 vizinhos para uma rede Hexagonal de 3 vizinhos ou uma Cúbica de 6 vizinhos) não muda a classe de universalidade do modelo, mas altera drasticamente os valores exatos da temperatura crítica Tc​. Além disso, em 2D, o modelo de Potts tem uma transição de fase contínua para q≤4 e de primeira ordem para q>4. Em 3D, a transição é de primeira ordem para q≥3. Ter os grafos já prontos para qualquer dimensão permite investigar essas mudanças facilmente.
A Vantagem Prática da Lista de Arestas

O formato de arquivo .txt em lista de arestas que criamos se encaixa como uma luva no cálculo da energia do modelo de Potts. O Hamiltoniano do modelo é dado por:

Onde o somatório ocorre sobre todos os pares de primeiros vizinhos ⟨i,j⟩, e δ é a função delta de Kronecker (vale 1 se os spins forem iguais, e 0 se forem diferentes).

Como os grafos já contêm toda a complexidade topológica embutida nas conexões, o seu algoritmo de simulação térmica fica extremamente enxuto e ignorante em relação à geometria: ele apenas segue os pares de nós.

Para a dinâmica local do Metropolis, existe um detalhe prático de implementação: embora a Lista de Arestas seja perfeita para salvar no disco e calcular a energia global, na hora do sorteio nós precisamos saber rapidamente quem são os vizinhos de um nó específico. Para isso, ao carregar o arquivo, nós o convertemos para uma Lista de Adjacência (uma lista de listas).

Abaixo está o código base do algoritmo de Metropolis para o modelo de Potts.
O Passo de Metropolis para o Modelo de Potts

A lógica para um passo (ou "varredura") de Monte Carlo, que consiste em N tentativas de giro, é a seguinte:

    Sorteie um nó aleatório da malha.

    Proponha um novo estado aleatório entre 0 e q−1 (diferente do atual).

    Calcule a variação de energia ΔE comparando os vizinhos diretos.

    Se ΔE≤0, aceite a mudança.

    Se ΔE>0, aceite apenas com probabilidade p=exp(−ΔE/T), onde consideramos a constante de Boltzmann kB​=1 e a constante de acoplamento J=1.

```python
import numpy as np

def carregar_vizinhos(caminho_arquivo, N):
    '''
    Lê a lista de arestas do .txt e cria uma Lista de Adjacência.
    N é o número total de nós da malha (ex: Lx * Ly).
    '''
    # skiprows=1 ignora o cabeçalho 'Source Target'
    arestas = np.loadtxt(caminho_arquivo, dtype=int, skiprows=1)

    vizinhos = [[] for _ in range(N)]
    for u, v in arestas:
        vizinhos[u].append(v)
        vizinhos[v].append(u)

    return vizinhos

def passo_metropolis(spins, vizinhos, q, T):
    '''
    Executa 1 Varredura de Monte Carlo (MCS).
    Tenta atualizar N nós sequencialmente.
    '''
    N = len(spins)

    for _ in range(N):
        # 1. Sorteia um nó aleatório
        i = np.random.randint(N)
        estado_atual = spins[i]

        # 2. Sorteia um novo estado (diferente do atual)
        novo_estado = np.random.randint(0, q)
        while novo_estado == estado_atual:
            novo_estado = np.random.randint(0, q)

        # 3. Calcula a variação de energia local (delta E)
        E_inicial = 0
        E_final = 0

        for viz in vizinhos[i]:
            if spins[viz] == estado_atual:
                E_inicial -= 1  # Conta vizinhos iguais ao estado antigo
            if spins[viz] == novo_estado:
                E_final -= 1    # Conta vizinhos iguais ao novo estado proposto

        delta_E = E_final - E_inicial

        # 4 e 5. Regra de aceitação de Metropolis
        if delta_E <= 0:
            spins[i] = novo_estado
        else:
            # Sorteia um número entre 0 e 1 para chance de aceitação térmica
            if np.random.rand() < np.exp(-delta_E / T):
                spins[i] = novo_estado

    return spins
```

Calcular a susceptibilidade magnética (junto com o calor específico) é fundamental, pois essas grandezas são derivadas das flutuações do sistema. Quando o sistema passa por uma transição de fase, as flutuações divergem, e é exatamente aí que você encontra a temperatura crítica Tc​.

Em simulações de Monte Carlo, não calculamos a susceptibilidade aplicando um campo magnético externo e derivando. Usamos o Teorema de Flutuação-Dissipação, que relaciona a susceptibilidade à variância da magnetização, e o calor específico à variância da energia.
A Matemática dos Observáveis no Modelo de Potts

1. Energia Total (E)
Basta somar as interações de todos os pares conectados na rede (a lista de arestas brilha muito aqui):
E=−⟨i,j⟩∑​δ(σi​,σj​)

2. Magnetização / Parâmetro de Ordem (M)
No modelo de Ising, a magnetização é só a média dos spins. No modelo de Potts (q estados), isso não funciona, pois não há "para cima" e "para baixo". A definição padrão do parâmetro de ordem macroscópico baseia-se na população do estado majoritário (o estado que tem mais spins):
M=N(q−1)qNmax​−N​

Onde Nmax​ é o número de spins no estado mais populoso e N é o número total de nós. Se todos os spins forem iguais, M=1. Se o sistema estiver totalmente desordenado (cada estado tem N/q spins), M=0.

3. Susceptibilidade Magnética (χ) e Calor Específico (Cv​)
Calculados após a termalização, tirando as médias das grandezas e de seus quadrados:
χ=TN​(⟨M2⟩−⟨M⟩2)
Cv​=NT21​(⟨E2⟩−⟨E⟩2)
Implementação em Python

Aqui estão as funções para medir esses valores em cada passo, e como integrá-las no laço principal da sua simulação.

```python
import numpy as np

def calcular_energia_total(spins, arestas):
    '''Calcula a energia total do sistema varrendo a lista de arestas.'''
    E = 0
    # Como arestas é um array (N_arestas, 2), iteramos sobre os pares
    for u, v in arestas:
        if spins[u] == spins[v]:
            E -= 1  # Interação ferromagnética (J=1)
    return E

def calcular_magnetizacao_potts(spins, q):
    '''Calcula o parâmetro de ordem M para o modelo de Potts.'''
    N = len(spins)
    # np.bincount é extremamente rápido para contar frequências de inteiros
    contagens = np.bincount(spins, minlength=q)
    N_max = np.max(contagens)

    M = (q * N_max - N) / ((q - 1) * N)
    return M

# ==============================================================================
# LAÇO PRINCIPAL DE SIMULAÇÃO (MEDIDAS)
# ==============================================================================

# Supondo que você já tem: spins, vizinhos, arestas, q, T, N
# E que a função passo_metropolis (do código anterior) está definida.

passos_termalizacao = 2000  # Passos jogados fora (para chegar ao equilíbrio)
passos_medida = 5000        # Passos usados para calcular as médias

# 1. Termalização
for _ in range(passos_termalizacao):
    spins = passo_metropolis(spins, vizinhos, q, T)

# 2. Medição
energia_acumulada = 0.0
energia_quad_acumulada = 0.0
mag_acumulada = 0.0
mag_quad_acumulada = 0.0

for _ in range(passos_medida):
    # Executa o passo de Monte Carlo
    spins = passo_metropolis(spins, vizinhos, q, T)

    # Mede os observáveis da configuração atual
    E_atual = calcular_energia_total(spins, arestas)
    M_atual = calcular_magnetizacao_potts(spins, q)

    # Acumula para as médias
    energia_acumulada += E_atual
    energia_quad_acumulada += E_atual**2

    mag_acumulada += M_atual
    mag_quad_acumulada += M_atual**2

# 3. Cálculos Finais (Médias Termodinâmicas)
E_media = energia_acumulada / passos_medida
E2_media = energia_quad_acumulada / passos_medida

M_media = mag_acumulada / passos_medida
M2_media = mag_quad_acumulada / passos_medida

# Usando o Teorema de Flutuação-Dissipação
Calor_Especifico = (E2_media - E_media**2) / (N * (T**2))
Susceptibilidade = (N / T) * (M2_media - M_media**2)

# Normalizando a energia por spin para facilitar a leitura
Energia_por_Spin = E_media / N

print(f"Resultados para T = {T}:")
print(f"Energia por Spin : {Energia_por_Spin:.4f}")
print(f"Magnetização (M) : {M_media:.4f}")
print(f"Calor Específico : {Calor_Especifico:.4f}")
print(f"Susceptibilidade : {Susceptibilidade:.4f}")
```

Para mapear a transição de fase perfeitamente, você criará um laço externo iterando sobre várias temperaturas. Por exemplo, criando um vetor de temperaturas (temperaturas = np.linspace(0.5, 2.5, 30)).

Para temperaturas altas, M≈0 e χ será baixo. Para temperaturas muito baixas, M≈1 e χ também será baixo (o sistema está congelado). Perto da Tc​, a susceptibilidade dará um "salto" gigantesco (um pico). Se você aplicar isso nas topologias torcidas (como a Fita de Möbius), verá que o pico de calor específico e susceptibilidade sofre um alargamento e um deslocamento em relação ao Toro perfeito, justamente por causa do defeito macroscópico forçado pela geometria da fronteira.

```python
import numpy as np
import os

def lista_arestas_cayley_fantasma(z, geracoes):
    '''
    Gera Árvore de Cayley com um Nodo Fantasma conectado a todas as folhas.
    Retorna a lista de arestas e o ID do nodo fantasma para ser tratado
    com um peso diferente na simulação.
    '''
    arestas = []
    folhas = []

    if geracoes == 0:
        return np.array(arestas), -1

    no_atual = 0
    proximo_no_livre = 1
    fila = [(no_atual, 0)]

    while fila:
        no, gen = fila.pop(0)

        # Se ainda não estamos na última camada, criamos os filhos
        if gen < geracoes:
            num_filhos = z if no == 0 else z - 1
            for _ in range(num_filhos):
                arestas.append((no, proximo_no_livre))
                fila.append((proximo_no_livre, gen + 1))
                proximo_no_livre += 1
        else:
            # Se chegou na última geração (gen == geracoes), é uma folha!
            folhas.append(no)

    # ---------------------------------------------------------
    # TRUQUE DO MESTRADO: O NODO FANTASMA
    # ---------------------------------------------------------
    id_nodo_fantasma = proximo_no_livre

    for folha in folhas:
        # Conecta a folha ao nodo fantasma
        arestas.append((folha, id_nodo_fantasma))

    return np.array(arestas), id_nodo_fantasma

# ==============================================================================
# PAINEL DE EXECUÇÃO
# ==============================================================================

if __name__ == "__main__":
    z = 3
    geracoes = 5

    pasta_saida = "listas_arestas_hierarquicas"
    if not os.path.exists(pasta_saida):
        os.makedirs(pasta_saida)

    print(f"Gerando Árvore de Cayley com Nodo Fantasma (z={z}, gen={geracoes})...")

    arestas, id_fantasma = lista_arestas_cayley_fantasma(z, geracoes)

    nome_arquivo = f"Cayley_Fantasma_z{z}_gen{geracoes}.txt"
    caminho_completo = os.path.join(pasta_saida, nome_arquivo)

    # Salva a lista de arestas padrão
    np.savetxt(caminho_completo, arestas, fmt='%d', delimiter=' ',
               header='Source Target', comments='')

    print(f" -> Rede gerada com sucesso!")
    print(f" -> Total de Nós: {id_fantasma + 1} (O Nodo Fantasma é o ID: {id_fantasma})")
    print(f" -> Total de Arestas: {len(arestas)}")
```
